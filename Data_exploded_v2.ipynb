```python
import pandas as pd
```

```python
df = pd.read_csv('//content//marketing_sample_for_amazon_com-ecommerce__20200101_20200131__10k_data.csv')
```

```python
#from Bruna
columns_to_keep = ['Uniq Id', 'Product Name', 'Category', 'Upc Ean Code', 'Selling Price', 'Model Number',
                   'About Product', 'Product Specification', 'Technical Details', 'Shipping Weight',
                   'Product Dimensions', 'Image', 'Variants', 'Product Url', 'Is Amazon Seller']
df = df[columns_to_keep].copy()
```

```python
exploded_df = df.assign(Image=df['Image'].str.split('|')).explode('Image')
```

```python
exploded_df.head()
```

```python
exploded_df.fillna('', inplace=True)
```

```python
exploded_df.to_csv('output.csv', index=False)
```

```python
columns_to_concatenate = {
    'Product Name': 'This product is named',
    'Category': 'It falls under the category',
    'Upc Ean Code': 'The UPC EAN code is',
    'Selling Price': 'The selling price is',
    'Model Number': 'The model number is',
    'About Product': 'About the product:',
    'Product Specification': 'Product specifications:',
    'Technical Details': 'Technical details:',
    'Shipping Weight': 'The shipping weight is',
    'Product Dimensions': 'The product dimensions are',
    'Variants': 'Variants available at',
    'Product Url': 'More details can be found at',
    'Is Amazon Seller': 'Is it an Amazon Seller:',
}

# Convert dictionary keys to a list for indexing
cols_to_fill = list(columns_to_concatenate.keys())

# Fill NaN values with empty strings to avoid errors during concatenation
exploded_df[cols_to_fill] = exploded_df[cols_to_fill].fillna('')

# Concatenate the columns with introductory phrases
exploded_df['Text'] = df.apply(lambda row: ' '.join([f"{phrase} {row[col]}" for col, phrase in columns_to_concatenate.items()]), axis=1)

```

```python
text_df = exploded_df[['Uniq Id', 'Text']].drop_duplicates()
```
